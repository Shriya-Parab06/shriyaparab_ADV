```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset to examine its structure
file_path = '/content/bodyfat.csv'
df = pd.read_csv(file_path)
df
```

```python
# Display the first few rows of the dataset and the summary to understand its structure
df.head(), df.describe(), df.info()
```

```python
import plotly.express as px

# Verify the column names in your DataFrame
print(df.columns)

# Choose existing columns for the bar chart
bar_chart = px.bar(df, x='Age', y='BodyFat', title='Body Fat Percentage by Age')
bar_chart.show()
```

Observations:

Peak in Mid-40s: The chart shows a clear peak in body fat percentage around the age of 45. This suggests that body fat tends to accumulate significantly in the mid-40s.

Gradual Decline: After the peak, there's a general trend of decreasing body fat percentage with increasing age. This could be attributed to various factors, including reduced muscle mass and metabolic rate.

```python
# Pie chart based on 'Age' and 'Weight'
pie_chart = px.pie(df, names='Age', values='Weight', title='Distribution of Weight Percentage by Age Group')
pie_chart.show()
```

Observations:

Age 40 Dominates: The largest segment of the pie chart is represented by the age group 40, indicating that it has the highest weight percentage distribution among the groups shown.

Clustering Around 40: Several age groups in the vicinity of 40, such as 41, 42, 43, and 44, also have relatively large segments, suggesting a clustering of weight percentage in this age range.

```python
# Histogram
histogram = px.histogram(df, x='Density', title='Distribution of Density')
histogram.show()
```

Observations:

Central Tendency: The peak of the curve appears to be around 1.05, suggesting that this is the most common density value in the dataset.

Spread: The spread of the data, as measured by the width of the curve, appears to be moderate. This indicates that the data points are not too tightly clustered around the central value but are also not widely dispersed.

Skewness: The curve appears to be slightly skewed to the right, meaning there is a longer tail extending to the right side of the distribution. This suggests that there are a few data points with higher density values that are pulling the mean to the right.

Outliers: There are a few outliers visible in the tails of the distribution, particularly on the left side. These are data points that are significantly different from the majority of the data and may require further investigation.

```python
# Scatter Plot (Using 'Weight' and 'BodyFat')
scatter_plot = px.scatter(df, x='Weight', y='BodyFat', color='Age',
                         title='Body Fat vs Weight, Colored by Age',
                         labels={'Weight': 'Weight (lbs)', 'BodyFat': 'Body Fat %'})
scatter_plot.show()
```

Observations:

Positive Correlation: There appears to be a general positive correlation between body fat percentage and weight. This suggests that as weight increases, body fat percentage tends to increase as well.

```python
# Bubble Plot (Using 'Weight', 'BodyFat', and 'Age' for bubble size)
bubble_plot = px.scatter(df, x='Height', y='BodyFat', size='Age', color='Age',
                        title='Body Fat vs height, Bubble Size by Age',
                        labels={'Height': 'Height (inches)', 'BodyFat': 'Body Fat %'})
bubble_plot.show()
```

Observations:

Weak Positive Correlation: There seems to be a weak positive correlation between body fat percentage and height, though it's not as strong as in the previous plot. This suggests that taller individuals tend to have slightly higher body fat percentages, but the relationship is not particularly strong.

```python
# Box Plot
box_plot = px.box(df, x='Age', y='Neck', title='Neck circumference Distribution by Age')
box_plot.show()
```

Observations:

Median: The median neck circumference increases gradually with age. For example, the median for the age group 30-35 is lower than the median for the age group 70-75.

Interquartile Range (IQR): The IQR, which represents the middle 50% of the data, doesn't show a significant change with age. This indicates that the spread of neck circumferences within each age group is relatively consistent.

```python
# Whisker Plot (Box plot is essentially a type of whisker plot)
whisker_plot = px.box(df, x='Age', y='Abdomen', points='all', title='Abdomen Circumference Distribution by Age (with all points)')
whisker_plot.show()
```

Observations:

Median: The median abdomen circumference increases gradually with age. For example, the median for the age group 30-35 is lower than the median for the age group 70-75.

Interquartile Range (IQR): The IQR, which represents the middle 50% of the data, shows a slight increase with age. This indicates that the spread of abdomen circumferences within each age group becomes slightly wider as people get older.

```python
#  violin plot
# Select data for a specific age range \
df_filtered = df[(df['Age'] >= 40) & (df['Age'] <= 50)]

# Create the violin plot
violin_plot = px.violin(df_filtered, y='BodyFat', box=True, points='all',
                        title='Body Fat Distribution for Ages 40-50')
violin_plot.show()
```

Observations:

Median Around 20: The median body fat percentage for the age group 40-50 appears to be around 20%.

Skewness: The distribution is slightly skewed to the right, indicating that there are a few individuals with higher body fat percentages that are pulling the mean to the right.

```python
# Linear Regression Plot
linear_regression = px.scatter(df, x='Weight', y='BodyFat', trendline='ols',
                               title='Linear Regression: Body Fat vs Weight')
linear_regression.show()
```

Observations:

Positive Linear Relationship: The scatter plot shows a clear positive linear relationship between body fat percentage and weight. This indicates that as weight increases, body fat percentage tends to increase as well.

```python
# 3D Chart
fig_3d = px.scatter_3d(df, x='Biceps', y='Forearm', z='Wrist', color='BodyFat',
                      title='3D Scatter: Biceps vs Forearm vs Wrist ')
fig_3d.show()
```

Observations:

No Strong Linear Relationships: There doesn't appear to be a strong linear relationship between body fat percentage and any of the three measurements (biceps, forearm, or wrist). This is evident from the scattered nature of the data points in the 3D space.

```python
# Donut Chart
donut_chart = px.pie(df, names='Age', values='Height', hole=0.4,
                     title='Distribution of Height Percentage by Age Group (Donut)')
donut_chart.show()
```

Observations:

Age 36 Dominates: The largest segment of the donut chart is represented by the age group 36, indicating that it has the highest height percentage distribution among the groups shown.

Clustering Around 36: Several age groups in the vicinity of 36, such as 38, 39, and 40, also have relatively large segments, suggesting a clustering of height percentage in this age range.

```python
# Waterfall Chart
import plotly.graph_objects as go
fig_waterfall = go.Figure(go.Waterfall(
    name = "BodyFat Changes",
    orientation = "v",
    measure = ["absolute", "relative", "total"],
    x = ["Initial", "Exercise", "Diet", "Final"],
    textposition = "outside",
    text = ["+15%", "-20%", "-10%", "Total"],
    y = [15, -10, -5, 10],
    connector = {"line":{"color":"rgb(63, 63, 63)"}},
))

fig_waterfall.update_layout(
    title = "BodyFat Changes Waterfall Chart",
)

fig_waterfall.show()
```

```python
# Treemap
!pip install squarify
import squarify
sizes = df['BodyFat'].value_counts()
labels = sizes.index
squarify.plot(sizes=sizes, label=labels, alpha=.8 )
plt.axis('off')
plt.title('Body Fat Distribution Treemap')
plt.show()
```

Observations:

Smaller Clusters: Smaller clusters can be seen in the upper and lower corners of the treemap, representing individuals with lower or higher body fat percentages.

Outliers: A few outliers, such as the very small green square in the upper left corner, represent individuals with extremely low body fat percentages.
